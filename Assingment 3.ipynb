### 1) Plot a scatter plot for x = [5,7,8,7,2,17,2,9,4,11,12,9,6] and y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y1 = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x1, y1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### 2) Plot a scatter plot for log(x) for x= 1 to 10

```python
x2 = [ i for i in range(1,11)]
y2 = np.log10(x2) 
plt.scatter(x2, y2)
plt.xlabel("x")
plt.ylabel("y=log10(x)")
plt.show()
```
